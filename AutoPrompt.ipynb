```python
# Parte 1: Simular construcción de prompt para chatbot

# Prompt del sistema (prompt.txt)
system_prompt = """
Eres un asistente educativo que responde de forma amigable, clara y sin opiniones políticas.
"""

# Historial de conversación (usuario y asistente)
conversation_history = [
    {"role": "user", "content": "Hola, ¿qué es la fotosíntesis?"},
    {"role": "assistant", "content": "La fotosíntesis es el proceso mediante el cual las plantas convierten la energía solar en alimento."}
]

# Nuevo mensaje del usuario
new_user_message = {"role": "user", "content": "¿Y qué organismos la realizan?"}

# Construir el "payload" completo
messages = []

# Primero el prompt de sistema
messages.append({"role": "system", "content": system_prompt.strip()})

# Luego el historial
messages.extend(conversation_history)

# Finalmente el nuevo mensaje del usuario
messages.append(new_user_message)

# Mostrar el "payload" listo para enviar a un modelo (como gpt-3.5-turbo)
for msg in messages:
    print(f"[{msg['role'].upper()}]: {msg['content']}\n")

# ---

# Parte 2: Simular Automatic Prompt Optimization

from random import choice

# Prompt inicial para la tarea
prompt_inicial = "Resume el siguiente texto en 20 palabras."

# Variantes automáticas simples
variantes_prompt = [
    "Haz un resumen breve (20 palabras máximo) del siguiente contenido.",
    "Sintetiza el texto siguiente usando unas 20 palabras.",
    "Escribe una breve sinopsis de aproximadamente 20 palabras del siguiente texto."
]

# Texto de prueba
texto_prueba = """
La fotosíntesis es un proceso mediante el cual las plantas, algas y algunas bacterias convierten la energía solar en energía química.
"""

# Función ficticia para "simular" respuestas de un modelo
# (En la vida real aquí llamaríamos al modelo para generar respuestas)
def simular_respuesta(prompt):
    if "breve" in prompt:
        return "Las plantas convierten luz solar en energía química mediante la fotosíntesis."
    elif "sintetiza" in prompt:
        return "Proceso donde las plantas usan luz solar para generar energía en forma de glucosa."
    else:
        return "Las plantas, algas y bacterias transforman la energía solar en energía química mediante la fotosíntesis."

# Evaluar cada variante
mejor_prompt = None
mejor_respuesta = None
mayor_calificacion = -1

for variante in variantes_prompt:
    respuesta = simular_respuesta(variante)
    longitud = len(respuesta.split())  # Simple métrica: queremos que la respuesta tenga entre 18 y 22 palabras
    
    # Calificar según qué tan cerca está de 20 palabras
    puntuacion = max(0, 20 - abs(20 - longitud))

    print(f"Prompt: {variante}")
    print(f"Respuesta: {respuesta}")
    print(f"Puntuación: {puntuacion}\n")

    if puntuacion > mayor_calificacion:
        mejor_prompt = variante
        mejor_respuesta = respuesta
        mayor_calificacion = puntuacion

# Mostrar el mejor prompt encontrado
print("==== Mejor Prompt Encontrado ====")
print(f"Prompt: {mejor_prompt}")
print(f"Respuesta generada: {mejor_respuesta}")
print(f"Puntuación: {mayor_calificacion}")
```
